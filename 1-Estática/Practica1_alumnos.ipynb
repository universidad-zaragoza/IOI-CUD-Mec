<h2><b>SISTEMA ESTUDIADO.</b></h2>



&emsp; El disco de embrague, en negro en la figura, gira solidariamente con la caja de cambios (que forma parte de la transmisión que va conectada a las ruedas tractoras) y está embutido entre el volante de inercia y el plato de presión que giran solidariamente con el motor.<br>

&emsp; Cuando el sistema está embragado, el pedal queda en reposo y hay una serie de muelles helicoidales enrollados que están comprimidos y que, al intentar expandirse, aplastan el plato de presión contra el disco de embrague y éste a su vez contra el volante de inercia. De esta forma, a través del disco de embrague, el giro del bloque motor se transmite al sistema de la transmisión. Los muelles ejercen una fuerza perpendicular a las superficies de contacto implicadas y sobre ellas surge una fuerza de resistencia estática que toma el valor necesario para que el bloque motor y la caja de cambios giren al unísono (no deslicen).

&emsp; Cuando el sistema está desembragado, el pedal se pisa a fondo y a través de un juego de palancas se comprimen aún más los muelles helicoidales. Al acortar su longitud, el plato de presión con el que son solidarios se separa del disco de embrague que queda libre del bloque motor. Como resultado el motor y la caja de cambios se desacoplan y cada sistema puede girar sin problemas a velocidades angulares distintas. En
términos físicos: las fuerzas de contacto entre el disco de embrague y el bloque motor desaparecen.<br>

&emsp; En el modo de operación a "medio embrague" que puede ser de utilidad al arrancar para evitar que el sistema se "cale" (se quede bloqueado al no poder absorber las ruedas tractoras toda la potencia recibida) o que las ruedas patinen (al aumentar la fuerza tractora que ejercen por encima de lo permitido por su rozamiento con el terreno), el pedal de embrague no se pisa a fondo y la conexión entre el motor y la transmisión no se pierde. Sin embargo, sí que se transmite una fuerza a los muelles que intenta comprimirlos más aún, aunque no lo consiga. El resultado es que, aunque no se pierda el contacto, disminuye la fuerza normal con la que los muelles aprisionan al sistema de modo perpendicular y, con ella, el valor límite umbral de la fuerza de rozamiento por encima del cual empieza el deslizamiento entre el bloque motor y la transmisión. La idea es operar por debajo de ese límite para permitir que las ruedas empiecen a girar con una velocidad menor de la que tendrían con el sistema completamente embragado e ir soltando poco a poco el pedal para permitir un enganche suave entre el bloque motor y el sistema de la transmisión. En este modo de operación las piezas deslizan entre sí lo que las somete a desgaste.<br>
